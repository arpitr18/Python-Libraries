```python
df = sns.load_dataset('tips')
```

```python
df['size'].unique()
```

## Distribution Plot

```python
sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'],kde=True)
```

```python
sns.displot(df['total_bill'],bins=100)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'] , kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'] , kde=True)
```

## Categorical plot

```python
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'])
```

```python
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

```python
sns.boxplot(x= 'tip',y = 'day',data = df,palette = 'rainbow')
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
#The swarmplot is similar to stripplot(), but the points are adjusted 
#(only along the categorical axis) so that they dont overlap. 
#This gives a better representation of the distribution of values,
```

## Matrix Plot

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset("tips")
```

```python
tipscorr  = tips[['total_bill','tip','size']]
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot= True)
```

```python
sns.clustermap(tipscorr.corr())
```

```python
pvtflight = flights.pivot_table(values = 'passengers',index = 'month',columns = 'year')
```

```python
pvtflight
```

```python
sns.heatmap(pvtflight)
```

## Regression PLot

```python
tips = sns.load_dataset("tips")
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':50}) # scatter size is default 50 , you can change to any
```
